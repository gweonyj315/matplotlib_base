### 2. 축

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows(한글말 설정)
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # '-' 출력 정상화

x = [1, 2, 3]
y = [2, 4, 8]
```

```python
# 개별 폰트 설정

plt.plot(x, y)
plt.title('꺽은선 그래프', fontdict = {'family' : 'HYGungSo-Bold', 'size' : 20})
```

```python
# 축 범례 이름, 색깔, 위치 설정

plt.plot(x, y)
plt.xlabel('x축', color = 'red', loc = 'right') # left, center, right
plt.ylabel('y축', color = '#00aa00', loc = 'top') # top, center, bottom
```

```python
# 축 간격 설정

plt.plot(x, y)
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])
plt.show()
```
