# Matplotlib

다양한 형태의 그래프를 통해서 데이터를 시각화할 수 있는 라이브러리 

### 1. 그래프의 기본

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows(한글말 설정)
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # '-' 출력 정상화

x = [1, 2, 3]
y = [2, 4, 8]
plt.plot(x, y)
plt.show() # 그래프 위에 나오는 글자 없앨 수 있음

# Title 설정
plt.title('Line Graph') # 그래프의 이름 설정(원래는 한글 사용 x)
```

```python
import matplotlib.font_manager as fm # 폰트 확인 라이브러리 

fm.fontManager.ttflist # 사용가능한 폰트 확인
[f.name for f in fm.fontManager.ttflist] # 폰트 이름만 출력
```
